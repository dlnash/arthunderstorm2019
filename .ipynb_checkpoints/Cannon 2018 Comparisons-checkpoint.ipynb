```python
import logging
logging.captureWarnings(True)
logging.getLogger('py.warnings').setLevel(logging.ERROR)

import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import pandas as pd
from IPython.display import Image, display
from dask.distributed import Client, progress
from datetime import datetime, timedelta
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from cartopy.mpl.gridliner import LONGITUDE_FORMATTER, LATITUDE_FORMATTER
import matplotlib.ticker as mticker
import matplotlib.path as mpath
import matplotlib.patheffects as PathEffects
from scipy import stats
import scipy
import numpy.ma as ma
from mpl_toolkits.axes_grid1 import make_axes_locatable
from mpl_toolkits.axes_grid1 import ImageGrid
from scipy.ndimage import gaussian_filter
import metpy.calc as mpcalc
from metpy.units import units

from modules.plotter import make_cmap
# %matplotlib inline
```

```python
filename_pattern =  '/home/sbarc/students/nash/data/CFSv2/netcdf_CFSv2_vertical_levels/20140306cdas1.t06z.nc'
ds = xr.open_dataset(filename_pattern, engine='netcdf4')
print('ds size in GB {:0.2f}\n'.format(ds.nbytes / 1e9))
ds.info()
```

```python
da_SPFH = ds.spfh.sel(p=slice(1000,300))
da_V = ds.v.sel(p=slice(1000,300))
da_U = ds.u.sel(p=slice(1000,300))
pressure = ds.p.sel(p=slice(1000,300)).values

da_MSL = ds.prmsl
da_MSL = da_MSL.squeeze()/100 # convert to hPa

U_850 = ds.u.sel(p=850.0)* 1.94384 ## convert to knots
V_850 = ds.v.sel(p=850.0)* 1.94384 ## convert to knots

da_GPH = ds.hgtprs.sel(p=500.0).squeeze()

u_250 = ds.u.sel(p=250.0)
v_250 = ds.v.sel(p=250.0)

```

```python
lats = ds['latitude']
lons = ds['longitude']
```

### Calculate IVT

```python
g = -9.81 # gravity constant
ivt_u = scipy.integrate.trapz(da_U*da_SPFH, pressure, axis=1)*g
print(ivt_u.shape)
ivt_v = scipy.integrate.trapz(da_V*da_SPFH, pressure, axis=1)*g
print(ivt_v.shape)
ivt_total = np.sqrt(ivt_u**2 + ivt_v**2)
ivt_total = ivt_total.squeeze()
print(ivt_total.shape)

# iwv = scipy.integrate.trapz(da_SPFH, pressure, axis=1)*g
```

### Calculate 1000-850 EPT (k/hPa)

```python
def calc_ept(T, RH, p):
    '''Calculate the dewpoint from relative humidity and temperature
    then calculate EPT from T and Td''' 
    td = 243.5*(np.log(RH/100)+((17.67*T)/(243.5+T)))/(17.67-np.log(RH/100)-((17.67*T)/(243.5+T)))
    td = td.values +273.15 * units.degK
    T = T.values +273.15 * units.degK

    pres = np.tile(p, T.shape)
    pres = pres * units.hPa

    ept = mpcalc.equivalent_potential_temperature(pressure=p* units.hPa, temperature=T, dewpoint=td)
    
    return ept
```

```python
T850 = (ds.temp.sel(p=850.0) - 273.15) * units.degC ## convert to Celsius for dew point calculation
RH850 = ds.rh.sel(p=850.0)

T1000 = (ds.temp.sel(p=1000.0) - 273.15) * units.degC ## convert to Celsius for dew point calculation
RH1000 = ds.rh.sel(p=1000.0)

ept850 = calc_ept(T850, RH850, p=850.0)
ept1000 = calc_ept(T1000, RH1000, p=1000.0)

ept1000_850 = (ept1000-ept850)/(1000.0-850.0)
ept1000_850 = ept1000_850.squeeze()
print(ept1000_850.shape)
```

### Convert Omega to vertical velocity (m/s)

```python
# Load relevant variables
omega = ds.omega.sel(p=600.0)
t600 = ds.temp.sel(p=600.0)
mixing = ds.spfh.sel(p=600.0)
p600 = 600.0 * units.hPa

## Convert VVel Omega (Pa/s) with respect to pressure to Vvel with respect for height (m/s)
VVEL = mpcalc.vertical_velocity(omega, p600, t600, mixing)
VVEL = VVEL.squeeze()
print(VVEL.shape)
```

### Calculate Frontogenesis

```python
t700 = ds.temp.sel(p=700.0)
u700 = ds.u.sel(p=700.0)
v700 = ds.v.sel(p=700.0)
dx = np.diff(lons)
dy = np.diff(lats)

### Calculate potential temperature using Poisson equation
theta700 = mpcalc.potential_temperature(pressure=700.0* units.hPa, temperature=t700)

### Calculate Frontogenesis form of Petterssen Frontogenesis
front700 = mpcalc.frontogenesis(theta700, u700, v700, dx, dy, dim_order='yx')
front700 = front700.squeeze()
front700.shape
```
